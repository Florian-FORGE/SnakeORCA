```python
import sys
c_path = "/home/fforge/Stage-IA3D/scripts/"
sys.path.append(f"{c_path}/mutations")
import mutate_with_rdm as mm

gs = GridSpec(nrows= 1 , ncols=1)
f = plt.figure(clear=True, figsize=(60, 33.75))

bins = 250

ax = f.add_subplot(gs[0, 0])

mutations = mm.read_mutations_from_BED("original.bed")
random_mutations = mm.generate_random_pos_mutations(mutations=mutations, rdm_seed=3, distrib="rotate_bins")

l_mut1 = [[mut.start, mut.end] for mut in mutations]
l_mut2 = [[mut.start, mut.end] for mut in random_mutations]

print(l_mut1)
print(l_mut2)

bin_size = 128_000

prop1, prop2 = [], []
for bin_idx in range(250):
    bin_s = bin_idx * bin_size
    bin_e = bin_s + bin_size - 1
    
    overlap = 0
    for mut_start, mut_end in l_mut1:
        ov_start = max(bin_s, mut_start)
        ov_end = min(bin_e, mut_end)
        if ov_start <= ov_end:
            overlap += (ov_end - ov_start + 1)
    prop1.append(overlap / bin_size)

    overlap = 0
    for mut_start, mut_end in l_mut2:
        ov_start = max(bin_s, mut_start)
        ov_end = min(bin_e, mut_end)
        if ov_start <= ov_end:
            overlap += (ov_end - ov_start + 1)
    prop2.append(overlap / bin_size)


y = range(bins)
ax.barh(y, prop1, color="#9900A7", alpha=.7, edgecolor="black", linewidth=.01)
ax.barh(y, prop2, color="#472C49", alpha=.7, edgecolor="black", linewidth=.01)

ax.set_xlabel("Proportion Mutated", fontsize=18)
ax.set_ylabel("")
ax.set_yticks(ticks=[i for i in range(0, bins+1, bins//5)], labels=["0.0Mb", "6.3Mb", "12.7Mb", "19.1Mb", "25.5Mb", "31.9Mb"])
ax.tick_params(axis='both', labelsize=24)
ax.invert_yaxis()
```
